```python
import pandas as pd
data=pd.read_csv('E:/Data_Science/cook_book/Pandas-Cookbook-master\data\slb_stock.csv',index_col='Date')
```

```python
data_close=data['Close']
data_summary=data_close.describe(percentiles=[.1,.9])
upper_10=data_summary.loc['90%']
lower_10=data_summary.loc['10%']
criteria=((data_close < lower_10) | (data_close > upper_10))
data_top_bottom_10=data_close[criteria]
```

```python
import matplotlib.pyplot as plt
data_close.plot(color='black',figsize=(12,6))
data_top_bottom_10.plot(marker='o',style=' ',ms=4,color='red')

xmin=criteria.index[0]
xmax=criteria.index[-1]

plt.hlines(y=[lower_10,upper_10],xmin=xmin,xmax=xmax,color='black')
```

```python
data_close.plot(color='black',figsize=(12,6))
plt.hlines(y=[lower_10,upper_10],xmin=xmin,xmax=xmax,color='black')

plt.fill_between(x=criteria.index,y1=lower_10,y2=data_close.values,color='black')
plt.fill_between(x=criteria.index,y1=lower_10,y2=data_close.values,where=data_close<lower_10,color='red')
plt.fill_between(x=criteria.index,y1=upper_10,y2=data_close.values,where=data_close>upper_10,color='green')
```
